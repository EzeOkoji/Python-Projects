```python
"""
Question 2: Write a program (using functions!) that asks the user for a long string containing multiple words. Print 
back to the user the same string, except with the words in backwards order.
"""

user = input("Please enternter a sentence: ")

a = user.split(" ")

users = a[-1::-1]
print(users)
```

```python
"""
Question 3: Ask the user for a string and print out whether this string is a palindrome or not. (A palindrome is a 
string that reads the same forwards and backwards.)
"""

user_input = input("Please enter a word or number: ")

reverse = user_input[::-1]

if (user_input == reverse):
    print("Yes, this is a palindrome")
    
else:
    print("No, this is not a palindrome")


```
